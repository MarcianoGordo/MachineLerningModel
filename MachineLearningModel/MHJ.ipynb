```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split as tt
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import ExtraTreesRegressor
```

```python
data = pd.read_csv('C:/Users/joaoa/Documents/Projects/machine/Obesity Classification.csv')
data
```

```python
data ['Label'] = data['Label'].replace('Normal Weight',0)
```

```python
data ['Label'] = data['Label'].replace('Overweight',1)
```

```python
data ['Label'] = data['Label'].replace('Underweight',2)
```

```python
data ['Label'] = data['Label'].replace('Obese',3)
```

```python
data ['Gender'] = data['Gender'].replace('Male',22)
```

```python
data ['Gender'] = data['Gender'].replace('Female',11)
```

```python
y = data['Label']
x = data.drop('Label', axis = 1)
x
```

```python
x_tr,x_ts,y_tr,y_ts = tt (x,y, test_size = 0.3)
```
